```python
df = pd.read_csv("magic04.data",header=None)
```

```python
df = df.dropna()
```

```python
df[10].replace({'g': 1, 'h': -1}, inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
non_scaled_data_df = df.copy()
minmax_scaler = preprocessing.MinMaxScaler()
for column in list(df.columns):
    feature = df[column].to_numpy().reshape(-1, 1)
    scaled_feature = minmax_scaler.fit_transform(feature)
    df.loc[:, column] = scaled_feature
```

```python
df
```

```python
df[10].replace({1.0: 1, 0.0: -1}, inplace=True)
```

```python
df[10]=df[10].astype(int)
```
